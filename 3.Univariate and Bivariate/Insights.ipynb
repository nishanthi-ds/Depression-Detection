# Univariate and Bivariate

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load and Analyse Data

```python
df = pd.read_csv(r"C:\Nishanthi\Hope AI\Projects\Depression Detection\Ordered_Formate\2.Data- Preprocessing\Preprocessed_data.csv")
```

```python
df.head()
```

# Univariate Analysis

# 1.How many Male and Female present in the dataset?

```python
frequency = df['Gender'].value_counts().sort_index()
labels = {0: 'Female', 1: 'Male'}

for code, count in frequency.items():
    print(f"{labels[code]}: {count}")
```

# 2. What is the Average of Number of Days Indoor?

```python
mean = np.mean(df['Days_Indoors'])
print("Mean:", mean)
```

# 3. How many people have stress and not have stress?

```python
sns.histplot(df['Growing_Stress'])
plt.xlabel('Growing Stress')
plt.ylabel('Frequency')
plt.show()
```

# 4. How much Percentage of people have social weekness

```python
# Count 'Yes' and 'No' from the column
labels = ['No','Maybe','Yes']
sizes = [df['Social_Weakness'].value_counts()[0], df['Social_Weakness'].value_counts()[1],df['Social_Weakness'].value_counts()[2]]

# Plot the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()


```

# 5.Visualize Number of Days Indore

```python

plt.figure(figsize=(20, 8)) 
plt.scatter(df.index, df['Days_Indoors'], color='violet',s=1, alpha=0.9)
plt.xlabel('Index')
plt.ylabel('Days Indoors')
plt.show()

```

# Bivariate Analysis

# 6. What is the distribution of Growing Stress among Males and Females

```python
# stacked bar chart
gender_stress = pd.crosstab(df['Gender'], df['Growing_Stress'])
ax = gender_stress.plot(kind='bar', stacked=True, figsize=(8, 5), colormap='Set2')
ax.set_xticklabels(['Male', 'Female'])
plt.legend(['No','Yes'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

```

# 7. What is the relationship between each columns?

```python
# Plot correlation heatmap

corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(20, 20))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.tight_layout()
plt.show()
```

# 8. What is the distribution of Work Interest and Growing Stress


```python
# stacked bar chart
work_stress = pd.crosstab(df['Growing_Stress'], df['Work_Interest'])
ax = work_stress.plot(kind='bar', stacked=True, figsize=(8, 5), colormap='Set3')
ax.set_xticklabels(['No', 'Yes'])
plt.legend(['No','Maybe','Yes'], title='Work Interest')
plt.xlabel('Growing_Stress')
plt.ylabel('Count')
plt.show()
```
